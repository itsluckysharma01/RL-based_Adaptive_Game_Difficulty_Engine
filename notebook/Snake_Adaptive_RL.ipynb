# RL-based_Adaptive_Game_Difficulty_Engine

This repository contains an implementation of a Reinforcement Learning (RL) based adaptive game difficulty engine. The engine dynamically adjusts the difficulty level of a game based on the player's performance, providing a more engaging and personalized gaming experience.

“We first developed a baseline Snake game and modularized difficulty and metrics before introducing Reinforcement Learning.”

# Train DQN agent
python train.py

# Train PPO agent  
python train.py ppo

# Evaluate DQN agent
python evaluate.py dqn models/dqn_final.pth

# Evaluate PPO agent
python evaluate.py ppo models/ppo_final.pth
